```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
%matplotlib inline

plt.figure(figsize=(10,8))
days = [1,2,3,4,5]
labels = ["eat", "study", "code", "sleep"]
eat  =[2,2,1,3,1]
study=[8,6,9,7,5]
code =[6,8,9,7,6]
sleep=[8,8,5,7,12]

plt.stackplot(days, eat, study, code, sleep, labels=labels)

plt.legend()
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Stack Plot")
plt.grid(True)
```
