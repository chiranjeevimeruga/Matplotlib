## Overlapping Barchart with width function

```python
from matplotlib import pyplot as plt
%matplotlib inline

# By using %matplotlib inline we can eliminate plt.show() function
```

```python
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

population_china = [0.55, 0.61, 0.65, 0.72, 0.82, 0.92, 0.99, 1.07, 1.172, 1.239, 1.283]

population_india = [1.024, 1.139, 1.254, 1.268, 1.283, 1.337, 1.411, 1.455, 1.498, 1.551, 1.658]

population_us = [0.024, 0.139, 0.254, 0.268, 0.283, 0.337, 0.411, 0.455, 0.498, 0.551, 0.658]

plt.bar(year, population_us, label = 'US', Color = 'm', width=.8)
plt.bar(year, population_china, label = 'China', Color = 'b', width=.6)
plt.bar(year, population_india, label = 'India', Color = 'g', width=.4)

plt.xlabel('Year')
plt.ylabel('Population in Billion')
plt.legend()

# by using plt.legend() color indicator will be displayed
```

## Resizing Chart

```python
plt.figure(figsize=(10,7))
plt.bar(year, population_us, label = 'US', Color = 'm', width=.8)
plt.bar(year, population_china, label = 'China', Color = 'b', width=.6)
plt.bar(year, population_india, label = 'India', Color = 'g', width=.4)

plt.xlabel('Year')
plt.ylabel('Population in Billion')
plt.legend()
```
