```python
from matplotlib import pyplot as plt
import numpy as np

plt.style.use('classic')

year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

width = 0.25

indices = np.arange(len(year))

population_china = [1.397, 1.403, 1.409, 1.415, 1.42, 1.424, 1.428, 1.431, 1.434, 1.436, 1.438]

population_india = [1.324, 1.339, 1.354, 1.368, 1.383, 1.397, 1.411, 1.425, 1.438, 1.451, 1.458]

plt.bar(indices-width, population_china, width=width, label='China', color='red')
plt.bar(indices, population_india, width=width, label='India', color='green')

plt.xlabel('Year')
plt.ylabel('Population in Billion')
plt.title('Population of China vs India Comparision')
plt.legend()

plt.grid(True)
plt.xticks(ticks=indices, labels=year)
plt.show()
```
