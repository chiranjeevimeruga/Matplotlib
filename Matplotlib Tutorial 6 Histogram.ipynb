```python
from matplotlib import pyplot as plt

blood_sugar = [90,125,135,147,180,95,99,110,101,107,109,190,142,180]

bins = [80, 100, 125, 190]
# Diabetic_Blood_Sugar_Range

# 80-100 = normal
# 100-125 = pre diabetic
# above 125 = diabetic

plt.hist(blood_sugar, bins=3)
#rwidth seperates the histogram 
plt.show()
```

```python
#edgecolor for seperation between values
plt.hist(blood_sugar, bins=3, edgecolor='w')
plt.show()
```

```python
#rwidth function seperates the histogram with space indicated in value
plt.hist(blood_sugar, bins=3, rwidth=0.95) 
plt.show()
```

```python
# Changing Color of Histogram
plt.hist(blood_sugar, bins=3, rwidth=0.95, color='g')
plt.show()
```

```python
# Hollow Histogram
plt.hist(blood_sugar, bins=3, rwidth=0.95, color='g', histtype='step')
plt.show()
```

```python
from matplotlib import pyplot as plt

blood_sugar = [90,125,135,147,180,95,99,110,101,107,109,190,142,180]

blood_sugar_men = [90,125,135,147,180,95,99,110,101,107,109,190,142,180]
blood_sugar_women = [60,105,95,110,90,75,79,80,91,88,68,150,122,165]
bins = [80, 100, 125, 190]
type = [blood_sugar_men, blood_sugar_women]
colors=['c','r']
label = ['Men', 'Women']
# Diabetic_Blood_Sugar_Range

# 80-100 = normal
# 100-125 = pre diabetic
# above 125 = diabetic
plt.xlabel("Diabetic Blood Sugar Range")
plt.ylabel("Total No. of Patients")
plt.title("Blood Sugar Chart")
plt.hist(type, color=colors, bins=3, rwidth=0.95, label=label) 
plt.legend()
plt.show()
```

```python
# Changing Orientation
plt.hist(type, color=colors, bins=3, rwidth=0.95, label=label, orientation='horizontal') 
plt.legend()
plt.show()
```
